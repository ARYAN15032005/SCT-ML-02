```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#load data
df = pd.read_csv('Mall_Customers.csv')  
print("Dataset loaded successfully!")
print("\nFirst 5 rows:")
print(df.head())
```

```python
#Data Preprocessing
```

```python
# Check for missing values
print("\nMissing values check:")
print(df.isnull().sum())
```

```python
# Select relevant features for clustering
features = ['Annual Income (k$)', 'Spending Score (1-100)']
X = df[features]
```

```python
# Scale the features (important for K-means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\nFirst 5 scaled samples:")
print(X_scaled[:5])
```

```python
#Determine Optimal Clusters (Elbow Method)
```

```python
# Calculate WCSS for different cluster numbers
wcss = []
cluster_range = range(1, 11)

for i in cluster_range:
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the Elbow Curve
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, wcss, marker='o', linestyle='--', color='blue')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.xticks(cluster_range)
plt.grid(True)
plt.show()
```

```python
#Apply K-means Clustering
```

```python
# Based on the elbow plot, choose optimal clusters (let's assume 5)
optimal_clusters = 5

# Apply K-means with selected clusters
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(X_scaled)
```

```python
# Add cluster labels to original dataframe
df['Cluster'] = clusters

# Get cluster centers in original scale
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)

print("\nCluster centers (original scale):")
print(cluster_centers)
```

```python
# Create a beautiful cluster visualization
plt.figure(figsize=(7, 5))
sns.scatterplot(x=features[0], y=features[1], data=df, 
                hue='Cluster', palette='viridis', s=100)

# Plot cluster centers
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], 
            s=300, c='red', marker='X', label='Centroids')

# Add labels and title
plt.title('Customer Segmentation', fontsize=16)
plt.xlabel(features[0], fontsize=14)
plt.ylabel(features[1], fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Analyze each cluster's properties
cluster_summary = df.groupby('Cluster').agg({
    'Age': 'mean',
    'Annual Income (k$)': 'mean',
    'Spending Score (1-100)': 'mean',
    'Gender': lambda x: x.mode()[0],
    'CustomerID': 'count'
}).rename(columns={'CustomerID': 'Count'})

print("\nCluster Profile Summary:")
print(cluster_summary)
```

```python
# Visualize distributions
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Age distribution
sns.boxplot(x='Cluster', y='Age', data=df, ax=axes[0])
axes[0].set_title('Age Distribution by Cluster')

# Income distribution
sns.boxplot(x='Cluster', y='Annual Income (k$)', data=df, ax=axes[1])
axes[1].set_title('Income Distribution by Cluster')

# Spending distribution
sns.boxplot(x='Cluster', y='Spending Score (1-100)', data=df, ax=axes[2])
axes[2].set_title('Spending Distribution by Cluster')

plt.tight_layout()
plt.show()
```

```python
# Interpretation of clusters
cluster_names = {
    0: "High Income, Low Spenders",
    1: "Moderate Income, High Spenders",
    2: "Low Income, Low Spenders",
    3: "High Income, High Spenders",
    4: "Moderate Income, Low Spenders"
}

df['Segment'] = df['Cluster'].map(cluster_names)

print("\nRecommended Marketing Strategies:")
print("1. High Income, High Spenders: Target with premium/luxury products")
print("2. High Income, Low Spenders: Develop engagement strategies")
print("3. Moderate Income, High Spenders: Loyalty programs")
print("4. Low Income, Low Spenders: Value deals and discounts")
print("5. Moderate Income, Low Spenders: Cross-selling opportunities")
```
